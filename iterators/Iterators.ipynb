Я поменял файл с последовательностями, так как в исходном почему-то последовательности были несколькими строками

```python
with open("sequences.fasta") as file, open("seq_mod.fasta", "w") as result:
    first_name = file.readline()
    result.write(first_name)
    sequence = ""
    for line in file:
        if line[0:4] == ">Seq":
            sequence = sequence.replace("\n", "")
            result.write(sequence + "\n")
            result.write(line)
            sequence = ""
        else:
            sequence += line
    sequence = sequence.replace("\n", "")
    result.write(sequence)
```

Сделал генератор

```python
def fasta_id(filename):
    count = 0
    seq_id = None
    seq = None
    with open(filename) as file:
        for line in file:
            if count % 2 == 0:
                seq_id = line
                seq_id = seq_id.replace("\n", "")
            else:
                seq = line
            if seq_id is not None and seq is not None:
                yield (seq_id + " " + seq)
                seq_id = None
                seq = None
            count += 1
```

```python
generator = fasta_id("seq_mod.fasta")
```

```python
for seq_id in generator:
    print(seq_id)
```

Сделал класс

```python
class Read_sequence:
    
    def __init__(self, filename):
        self.filename = filename
        self.aminoacids = np.array(["G","A","V","I","L","P","S","T","C","M","D","N","E","Q","K","R","H","F","Y","W"])
    
    
    def mutate_sequence_by_acids(self, seq):
        aminoacid_idx = np.random.randint(0,20)
        seq = seq.replace(self.aminoacids[aminoacid_idx], self.aminoacids[aminoacid_idx - 1 ])
        return seq
    
    def mutate_sequence_by_deletion(self,seq):
        seq_length = len(seq)
        start = seq_length // 10
        stop = seq_length // 8
        seq = seq[:start] + seq[stop+1:]
        return seq
        
    
    def __iter__(self):
        while True:
            with open(self.filename) as file:
                count = 0
                for line in file:
                    if count % 2 == 1:
                        choice = np.random.randint(1,21)
                        if choice == 5:
                            line = self.mutate_sequence_by_acids(line)
                            print("Mutated by aminoacid line below:\n")
                        elif choice == 6:
                            line = self.mutate_sequence_by_deletion(line)
                            print("Mutated by deletion line below\n")
                        yield line
                    count += 1
```

```python
seqs = Read_sequence("seq_mod.fasta")
```

```python
count = 0
for seq in seqs:
    print(seq)
    if count > 600:
        break
    count += 1
```
