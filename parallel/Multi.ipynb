У меня не работал модуль multiprocessing, но оказалось, что в юпитере с этим модулем вообще проблемы. Поэтому я установил адаптированную для юпитера версию  multiprocess

```python
from Bio import SeqIO
import multiprocess 
import multiprocessing
import time
```

```python
def parse_sequence(record_id, record_seq):
    parser = {}
    for word in record_seq:
        if word in parser:
            parser[word] += 1
        else:
            parser[word] = 1
    print(f"Contig {record_id}:    ", end = "")
    parsered = []
    for key, value in zip(parser.keys(), parser.values()):
            parsered.append(f"{key}={value}")
    print(*parsered, sep = ", ")
    return 
```

Сначала вариант реализации в лоб (то, что первое в голову приходит)

```python
def parse_fasta(filename, threads):
    records = SeqIO.parse(filename, "fasta")
    selected_records = []
    for record in records:
        selected_records.append(record)
        if len(selected_records) == threads:
            processes = []
            for record in selected_records:
                processes.append(multiprocess.Process(target=parse_sequence, args=(record.id, record.seq)))
            for proc in processes:
                proc.start()
            for proc in processes:
                proc.join()
            selected_records = []
    if len(selected_records) > 0: # take last ones
        processes = []
        for record in selected_records:
            processes.append(multiprocess.Process(target=parse_sequence, args=(record.id, record.seq)))
        for proc in processes:
            proc.start()
        for proc in processes:
            proc.join()    
```

```python
start = time.time()
parse_fasta("contigs.fna", 1)
end = time.time()
print(f"executing time is {end-start}")
```

```python
start = time.time()
parse_fasta("contigs.fna", 2)
end = time.time()
print(f"executing time is {end-start}")
```

Теперь более сложная реализация

```python
class MyProcess(multiprocess.Process):
    def __init__(self, records):
        super().__init__()
        self.records = records
    
    def run(self):
        for record in self.records:
            parse_sequence(record.id, record.seq)
```

```python
def parse_fasta_fast(filename, threads):
    records = SeqIO.parse(filename, "fasta")
    processes = []
    for i in range(threads):
        processes.append(MyProcess(records))
    for proc in processes:
        proc.start()
    for proc in processes:
        proc.join()
```

```python
start = time.time()
parse_fasta_fast("contigs.fna", 2)
end = time.time()
print(f"executing time is {end-start}")
```

Интересно то, что данная реализация отрабатывает нормально, то есть здесь отрабатывает один и тот же итератор в 2 процессах, повторения разных записей нет. Это не совсем вяжется с теорией на лекции, но это работает правильно
